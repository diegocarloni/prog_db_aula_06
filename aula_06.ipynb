## Mapeando valores

```python
avaliacao_vinhos['escala_preco'] = avaliacao_vinhos.price.map(lambda p: p / media_preco)
```

## Apply
---
### Aplica uma função de acordo com os valores de uma linha ou coluna

```python
media_pontos = avaliacao_vinhos.points.mean()
```

```python
def classifica_vinho(row):
    row['classificacao'] = "Acima da média" if row.points > media_pontos else "Abaixo da média"
    return row
```

```python
avaliacao_vinhos.apply(classifica_vinho, axis='columns')
```

```python
def classifica_valor_vinho(row):
    row['boolean_class'] = 1 if row.points > media_pontos else 0
    return row
```

```python
avaliacao_vinhos.apply(classifica_valor_vinho, axis='columns')
```
